```python
import requests
def api(message):
    response = None
    filter_n = 0
    while response is None and filter_n < 3:
        try:
            #print(message)
            #print(requests.post("http://9.135.143.158:8998/chat1", json=message, headers=self.headers))
            response = requests.post("http://9.134.231.57:8998/chat1", json=message, headers=headers).json()

            #response = json.loads(response)
            assert 'choices' in response
#             for i in range(len(response['choices'])):
#                 assert(response['choices'][i]['finish_reason']=="stop")
#                 assert('content' in response['choices'][i]['message'])
            return response
        except Exception as e:
            filter_n += 1
            print(e)
#             print(message)
#             print(response)
            response = None
    return {"choices":[{"message":{"content":""}}]}
```

```python
def generate_text(prompt):

    if not isinstance(prompt,list):
        prompt = [prompt]
    #self.reset()
    message = \
    {
        'model': engine,
        'messages': [{"role": "system", "content":system_prompt }],
        'n': n,
        'temperature':temperature,
        'top_p' : 0.9,
    } 


    for sub_prompt in prompt:
        message['messages'].append({"role": "user", "content": sub_prompt})
        #print(message)
        result = api(message=message)
        message['messages'].append({"role": "assistant", "content": result["choices"][0]["message"]["content"]})

    answer = [item["message"]["content"] for item in result["choices"]]     
    history = message['messages']
    #self.reset()
    return answer
```

```python
import re
prompt = '''
You are given several Feedbacks and some Exist Clusters.

For each feedback, you should analyze it and thinking:

According to the issue stated in the feedback, can this feedback be classified into one UNIQUE existing cluster?
Output your evaluation in the following format:

    #thereason: your analysis here.
    [key problem conveyed in feedback i]
    [similarity of feedback i & cluster 1 and your explanation]
    [similarity of feedback i & cluster 2 and your explanation]
        ....
    [similarity of feedback i & cluster n and your explanation]
    
    - IF this feedback can be classified into UNIQUE existing cluster, format the cluster  by <Cluster> and </Cluster>. Please note: a feedback can only belong to one cluster.
    - IF IT IS NOT, create a new cluster for this feedback. format the new cluster by <New Cluster> [cluster description] </New Cluster>.

###Example

##All Exist Cluster

Cluster1 
the misunderstanding of objects with counting words: "a/an"
Cluster2
Missing the clarification of how to handle the object without specific number.

##Feedback

Feedback 0
It could be that the prompt is not explicit enough about how to handle sentences that include numbers. For example, it doesn't specify that "two ovens" should be counted as two objects, not one. This lack of clarity might be causing the model to count only the number of unique objects mentioned, rather than the total number of objects.

Feedback 1
The second reason could be that the prompt does not specify whether 'a' and 'an' should be counted as '1'. The confusion can arise due to the vague instruction "The objects may be of different types and some may be specifically mentioned in multiples (like 'two chickens')". This may result in the model not counting objects mentioned with 'a' or 'an' properly, leading to incorrect output.

Feedback 2
Misunderstanding of Task: The prompt may not be conveying the task clearly to the model. The model is supposed to count individual objects, including multiple instances of the same object. However, the prompt might be interpreted by the model as counting types of objects instead of individual instances, leading to an incorrect answer. This can be observed in all examples, where the output seems to represent a count of unique objects rather than a total count of all objects.


Answer:

-Feedback 0:
The issue described in this feedback is mainly about how to count the items with a specific number. 
For cluster 1: this cluster is mainly about the a/an object misunderstanding, it is inconsistent with the main problem described in Feedback 0.
For cluster 2: cluster 2 is abount the missing expression of how to handle the objects with specific number.Feedback 0 can't be cluster in this category.
so it cannot be classified into an unique existing cluster.
Now I will create a new cluster for Feedback 0:
<New Cluster>
Cluster3:Incorrect handling of the items with specific numbers, such as regard "two ovens" should count the total numbers ranther than the number of unique item type.
</New Cluster>

-Feedback 1:
this feedback describe the issue about the item with 'a' or 'an' should be counted a '1'
For cluster 1:The cluster is about the misunderstanding of a/an object counting, it is similar to the issue in Feedback 1.
so Feedback 1 can be categorized in Cluster 1,the answer is <Cluster> Cluster1 </Cluster>.

-Feedback 2:
this feedback emphasize the aim to task is to count the total number of items,rather than count the type of items. 
For cluster 1: The cluster is about the uncorrect counting of a/an object , it is inconsistent with the issues described in Feedback 2.
For cluster 2: This cluster is abount the missing clarification of counting method about the objects with specific number.Feedback 2 can't be cluster in this category.
For cluster 3: Cluster 3  describe the incorrect counting of objects with specific number, somehow it's similar to the Feedback 2.
After checking the existing cluster and the previously created cluster, I think Feedback 2 can be categorized in Cluster3,the answer is <Cluster> Cluster3 </Cluster>.


so it cannot be classified into the existing cluster.

Now you complete the following task:
 
##All Exist Cluster

[cluster]

##Feedback

[feedback]
Answer:

'''
cluster_template = '''
Cluster[index]
[cluster_desc]
'''
feedback_template='''
Feedback[index]
[feedback]
'''
# clusters = [
#     '''the misunderstanding  of counting words: "a/an"''',
#     '''Missing the clarification of how to handle the item without specific number.''',
#     '''Incorrect handling of the items with specific numbers ,such as regard "three apples" as single item.''',
#     '''The model has difficulty in understanding the number in words("two","three",etc)'''
# ]
clusters = [
    '''the misunderstanding  of counting words: "a/an"''',
]
def build_cluster_text(clusters,interval='\n'):
    text = ""
    for index, cluster_desc in enumerate(clusters):
        text += cluster_template.replace('[index]',str(index+1)).replace('[cluster_desc]',cluster_desc)
        #text += interval
    return text

def build_feedback_text(feedbacks,interval='\n'):
    text = ""
    for index, feedback in enumerate(feedbacks):
        text += feedback_template.replace('[index]',str(index+1)).replace('[feedback]',feedback)
        #text += interval
    return text  


ex_match_pattern = "<Cluster>(.*?)</Cluster>"
new_match_pattern = "<New Cluster>(.*?)</New Cluster>"


# feedback='''
# Feedback 0
# The first reason could be that the prompt might not be clear enough about the need to take into account the specific quantities mentioned before the objects. When the text says "five carrots" or "two beds," it seems like the model is only counting them as one item each, rather than five and two respectively. The phrasing of the prompt might need to emphasize that if a quantity is specified, that quantity should be used in the total count.
# Feedback 1
# The second reason might be due to the inherent difficulty of the task. The model needs to understand not only the English language but also the concept of counting and basic arithmetic. This involves an understanding of numerical expressions and their relation to physical quantities, which might be a challenge for the model.
# Feedback 2
# Thirdly, it seems like the model might be misunderstanding certain items as quantities. For example, in the second example, "a toaster" and "a stove" could be interpreted as quantities (like "a pair" or "a dozen") rather than as singular items. This is a more complex language processing error that could be causing the model to miscount the total number of items.
# Feedback 3
# 1. Lack of clarity in the prompt: The current prompt might be unclear or ambiguous for the classifier to comprehend accurately. It might not be interpreting the task of counting quantities correctly, especially when the quantities are listed before the item. Improving the instruction on how to count the quantities properly might resolve this issue.
# Feedback 4
# 2. Misunderstanding of the "one" default: In the prompt, it's stated that "If a quantity is not explicitly mentioned for an item, assume there is one of that item." However, the classifier may not be correctly applying this rule, leading to an undercount of items. It's essential to make sure the classifier understands to assign a quantity of one when a quantity isn't explicitly stated.
# Feedback 5
# 3. Handling multiple items: From the results, it seems like the classifier might be struggling to correctly handle sentences with multiple items and their respective quantities. This could be due to a limitation in its ability to recognize and separate different items and their quantities in the sentence structure used. More explicit instructions regarding handling multiple items could help.

# '''
import random

feedback_path = "/Users/wangxiaolei/Downloads/AutoPrompter/Feedback_result/object_counting/Llama/APO_feedbacks.txt"
with open(feedback_path,'r',encoding='utf-8') as f:
    lines = f.readlines()
feedbacks = []
for line in lines:
    if line == '\n':
        continue
    else:
        feedbacks.append(line.strip())

batch_size = 5
cluster_result = []
feedbacks = feedbacks[:]
random.shuffle(feedbacks)
print(f'feedback_length:{len(feedbacks)}')
cluster_dict = {}

for ind in range(0,len(feedbacks),batch_size):
    batch_feedbacks = feedbacks[ind:ind+batch_size]
    feedback_text = build_feedback_text(batch_feedbacks)
    cluster_text = build_cluster_text(clusters)

    query_prompt = prompt.replace('[cluster]',cluster_text).replace('[feedback]',feedback_text)
    #print(query_prompt)
    answer = generate_text(query_prompt)
    answer = answer[0]
    print(answer)
    
    feedback_handle_results = answer.split('-Feedback')[1:]

    for handle_result in feedback_handle_results:
        #print(handle_result)
        if '<Cluster>' in handle_result:
            try:
                temp = re.search(ex_match_pattern,handle_result,re.DOTALL).group(1)
                temp = temp.strip()
                
                if temp not in cluster_dict:
                    cluster_dict[temp] = []
                cluster_dict[temp].append(len(cluster_result))
                cluster_result.append(temp)
            except:
                cluster_result.append("None")
                print(temp)
            
        elif '<New Cluster>' in handle_result:
            try:
                temp = re.search(new_match_pattern,handle_result,re.DOTALL).group(1)
                #print(temp)
                temp_list = temp.strip().split(':')
                if temp_list[0].strip() not in cluster_dict:
                    cluster_dict[temp_list[0].strip()] = []
                cluster_dict[temp_list[0].strip()].append(len(cluster_result))
                cluster_result.append(temp_list[0].strip())
                clusters.append(temp_list[1].strip())
            except:
                #feedback_handle_results[ind] = temp
                
                
                cluster_result.append("None")
                print(temp)
        else:
            cluster_result.append("None")
            
count_dict = {}
for item in cluster_result:
    if item not in count_dict:
        count_dict[item] = 0
    count_dict[item] += 1
print(count_dict)
for k,v in count_dict.items():
    #print(k)
    #print(k.split('Cluster'))
    try:
        k = int(k.split('Cluster')[1].strip())
        print(f'{clusters[k-1]}:{v}')
    except:
        print(k)

for k,v_list in cluster_dict.items():
    try:
        k = int(k.split('Cluster')[1].strip())
        print(f'{clusters[k-1]}:')
        for index in v_list:
            print("feedback")
            print(feedbacks[index])
        print("\n\n")
    except:
        print(k)
# print(clusters)

# for fed,cluster in zip(feedbacks,cluster_result):
#     cluster = cluster.strip()
#     print('feedback:')
#     print(fed)
#     print('cluster:')
#     print(clusters[int(cluster[-1])-1])
    
```
